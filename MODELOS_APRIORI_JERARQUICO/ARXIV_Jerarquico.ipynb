CÓDIGO **JERÁRQUICO**

```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import fcluster

# Cargar el dataset de Mall Customers
df = pd.read_csv('Mall_Customers.csv')
print(df.head())  # Ver las primeras filas del dataset para confirmar que se ha cargado correctamente

# Selección de características para clustering
features = ['Annual Income (k$)', 'Spending Score (1-100)']
df_selected = df[features]  # Aquí seleccionamos solo las dos características que queremos usar para el clustering

# Escalar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)  # Escalamos los datos para que todas las características estén en la misma escala

# Aplicar el clustering jerárquico
Z = linkage(df_scaled, method='ward')  # El método de Ward minimiza la varianza dentro de los clusters

# Graficar el dendrograma
plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title("Dendrograma de Clustering Jerárquico - Mall Customers")
plt.xlabel("Clientes")
plt.ylabel("Distancia Euclidiana")
plt.show()  # Mostramos el dendrograma para visualizar las uniones de los clusters

# Elegir el número de clusters cortando el dendrograma
clusters = fcluster(Z, t=5, criterion='maxclust')  # Cortamos el dendrograma para obtener 5 clusters

# Visualizar los clusters en un gráfico de dispersión, coloreando según el cluster asignado
plt.figure(figsize=(12, 6))
scatter = plt.scatter(df_selected['Annual Income (k$)'], df_selected['Spending Score (1-100)'],
                      c=clusters, cmap='rainbow', label=clusters)

# Título y etiquetas del gráfico
plt.title("Distribución de Clusters Jerárquicos - Mall Customers")
plt.xlabel("Ingreso Anual (k$)")
plt.ylabel("Spending Score (1-100)")

# Agregar la leyenda
handles, labels = scatter.legend_elements()
plt.legend(handles, [f'Cluster {i}' for i in range(1, 6)], title="Clusters")

# Mostrar el gráfico
plt.colorbar(label='Cluster')
plt.show()
```
